```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Define the years for training, validation, and testing
train_start, train_end = 2000, 2014
val_start, val_end = 2015, 2017
test_start, test_end = 2018, 2019

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(12, 4))

# Draw rectangles for each period
# Training period
rect_train = patches.Rectangle((train_start, 0), train_end - train_start, 1, linewidth=1, edgecolor='none', facecolor='lightblue', alpha=0.7)
ax.add_patch(rect_train)
ax.text((train_start + train_end) / 2, 0.5, 'Training (2000-2014)', ha='center', va='center', color='black', weight='bold')

# Validation period
rect_val = patches.Rectangle((val_start, 0), val_end - val_start, 1, linewidth=1, edgecolor='none', facecolor='lightgreen', alpha=0.7)
ax.add_patch(rect_val)
# Adjusted text position for validation period
ax.text((val_start + val_end) / 2, 0.2, 'Validation (2015-2017)', ha='center', va='center', color='black', weight='bold')

# Testing period
rect_test = patches.Rectangle((test_start, 0), test_end - test_start, 1, linewidth=1, edgecolor='none', facecolor='salmon', alpha=0.7)
ax.add_patch(rect_test)
# Adjusted text position for testing period
ax.text((test_start + test_end) / 2, 0.8, 'Testing (2018-2019)', ha='center', va='center', color='black', weight='bold')


# Set plot limits and labels
ax.set_xlim(train_start - 1, test_end + 1)
ax.set_ylim(0, 1)
ax.set_yticks([]) # Hide y-axis ticks
ax.set_xlabel("Year")
ax.set_title("Time-Based Data Splitting for Prediction Horizon")

# Add vertical lines to mark the splits
ax.axvline(train_end, color='gray', linestyle='--', linewidth=1)
ax.axvline(val_end, color='gray', linestyle='--', linewidth=1)


# Add an arrow to indicate prediction direction
ax.annotate('Prediction Horizon',
            xy=(test_end + 0.5, 0.5), xycoords='data',
            xytext=(test_end + 2, 0.5), textcoords='data',
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=-0.2"),
            ha='left', va='center')


plt.show()
```
